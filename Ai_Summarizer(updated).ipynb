```python
import fitz
from transformers import pipeline
import gradio as gr

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_number in range(doc.page_count):
        page = doc[page_number]
        text += page.get_text()
    return text

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(pdf_file):
    text = extract_text_from_pdf(pdf_file.name)
    summary = summarizer(text, max_length=500, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    return summary[0]['summary_text']

interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.File(type="filepath", label="Upload PDF"),  # Use 'filepath' or 'binary'
    outputs=gr.Textbox(label="Summary"),
    title="PDF Text Summarizer",
    description="Upload a PDF file and get a summarized text using Hugging Face transformers."
)

interface.launch()
```
